## Convert the duration of the song from milliseconds to seconds 

```python
# Convertion from ms to secs
df_tracks["duration"] = df_tracks.duration_ms.apply(lambda x: round(x/1000))
# dropping column `duration_ms`
df_tracks.drop("duration_ms", inplace=True, axis=1) 
df_tracks.duration.head()
```

## Correlation Map

1. Create visualization: the correlation map
2. Drop three unwanted keys, mode and explicit columns
3. Apply the pearson correlation method

- Set figure size of correlation map to (14, 6).  
- Use `heatmap()` function to create the correlation map & set `annotation = True` which will write the data value in each cell.  
- Set `fmt = ".1g"` -> string formatting quotes used when adding annotations.  
- `cmap` stands for the colour map.  

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
# heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

## Sample Only 4 Percent of the Whole Dataset

```python
sample_df = df_tracks.sample(int(.004*len(df_tracks)))
sample_df.head()
```

Will use `regplot()` function in the seaborn library to draw the regression plot.

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title="Loudness vs. Energy Correlation")
```

There is a very high positive correlation between loudness and energy. As energy increases, loudness of the song also increases and similiarly, if the song's loudness decreases, the energy of the track also decreases.

## Create a regression plot between popularity and acousticness in the form of a regression line.

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="b").set(title = "Popularity vs. Acousticness Correlation")
```

The regression line is in a downward direction, indicating that the as acousticness of the song increases, the popularity decreases. Similarily, as popularity increases, acousticness decreases.

## Data analysis based on genres of the songs

```python
df_genre = pd.read_csv("data/SpotifyFeatures.csv")
df_genre.head()
```

Here we will:
- Plot duration of the songs w.r.t different genres using a horizontal barplot

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y="genre", x="duration_ms", data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

- Find top five genres by Popularity and plot a barplot for popularity vs genre

### The following code is wrong.

We cannot analyze the data using the summation of popularity by genre. The data will be bias towards the Indie Genre when deciding fifth place, this is caused by the larger volume of indie music on Spotify compared to Dance. Thus, summing popularity by genre will result in Indie being place higher, just by the greater count of songs in that genre. 

As we see in the table above, Indie has 9543 songs while Dance has 8701. Indie has 842 more songs than the Dance genre. 

We should be taking the average of popularity for each genre, in order to get the correct result of this anaylsis.

```python
# This is WRONG in analyzing the top 5 most popular genres.
famous[['genre', 'popularity']].groupby(by="genre").sum().sort_values("popularity", ascending=False).head(6)
```
